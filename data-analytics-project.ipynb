## Health Factors:

The categories within the dataset contain nutrition information in the following categories: Energy (kCal), Carbohydrates (g), Protein (g), Fiber (g), Sugar (g), Total Fat (g), Saturated Fat (g), Trans Fat (g), Cholesterol (mg), and Sodium (mg). Based on research I have done on health, on YouTube channels such as [Thomas DeLauer](https://www.youtube.com/@ThomasDeLauerOfficial) and [Dr. Eric Berg](https://www.youtube.com/@DrEricBergDC), the nutrition factors that should most closely be watched are Carbohydrates and Trans Fats. 

It is agreed upon by all health groups that Trans Fats are not healthy in the least. Trans Fats are chemically altered fats that the human body cannot properly metabolize. They are linked to many health diseases such as diabetes and metabolic disorders. Thus, Trans Fats should be avoided at all costs. 

Carbohydrates, on the other hand, can be okay depending on the context. You don't want carbs to make up the majority of your diet. The way in which carbs affect the human body depends heavily upon both the quality of the carb itself and the lifestyle of the person consuming the carbs. Sugar and bread are among the lowest quality carbs. Table sugar (half glucose and half fructose) has a glycemic index of 50), while white bread has a glycemic index of 70 or greater. The glycemic index is a measure of how much a carb spikes one's blood sugar. When blood sugar is spiked, insulin is also raised in order to negate the high blood sugar effects. Net Carbs is what you should actually pay attention to. Net Carbs is the carbs plus sugar minus fiber. This is because fiber is only metabolized by the microbiome and acts as a buffer to slow the glucose spike from the carbs and sugar. 

This leads to the discussion of the Standard American Diet. The Standard American Diet is one of the most unhealthy diets one could eat. It is high in grains and processed foods. The result of spiking insulin multiple times a day for a long time is insulin resistance. Insulin resistance leads to diabetes which causes a whole host of other health problems. The healthier way to eat carbs is to do so in moderation (not every day) and when eating carbs, exercise, or at least walk. Higher quality carbs should be chosen over lower quality ones. One way to determine the quality of a carb is the glycemic index. The lower the glycemic index, the less it spikes your blood sugar. The glycemic load is another important factor in this consideration. In addition to these factors, natural carbs (not processed) such as berries and fruits are always the healthier choice since you get antioxidants in these which combat the oxidative damage of the sugars. 

When on a low-carb diet, one must eat higher amounts of fat for energy. Fat is more energy dense than carbs, so it is not necessary to eat as much in volume. Fat is much easier to store as body fat if eaten in excess than carbs. Carbs, if not used by the muscles for energy, trigger a mechanism known as de novo lipogenesis (DNL). DNL is the creation of new fat in the body. This requires a further explanation on sugar types. Glucose can be used by the muscles, however fructose is only metabolized by the liver and triggers DNL once eaten. DNL also halts any fat burning within the body. This is yet another problem with the Standard American Diet. So much of the food in the US has lots of added sugars and sweeteners. The most prominent sweetener is High Fructose Corn Syrup (HFCS), which is incredibly unhealthy. 

One more factor that requires explanation is the consumption of carbs and fats at the same time. In the history of humans, the available food sources would have been either high amounts of carbs (think fruits) or high amounts of fat (animal fat, nuts, fish, etc.), but not both at the same time. Research shows that eating high amounts of carbs and high amounts of fat at the same time can be addictive, which can lead to overeating. Thus, the consumption of high fat and high carb foods at the same time should be avoided.

### Explanation: Fat Types
It should be noted that products high in fat and low in carbs are not necessarily healthy options. There are different types of fats, some of which are healthier than others. The healthiest fat is monounsaturated fats. This type of fat can be found in Extra-Virgin Olive Oil, Avacado Oil, Avacados, and Macadamia Nuts. There are also saturated fats which, from my understanding, seems to be a neutral type of fat in terms of health. Then there are polyunsaturated fats. These are unstable fats and unfortunately are the most widely used in cooking, especially at fast food restaurants. The reason they are so widely used is because they are cheap. Polyunsaturated fats are highly present in oils like Canola Oil, Grapeseed Oil, Sunflower/Safflower Oils, and most all seed oils. The main problem with these oils is that they are very rancid, especially if they are re-used multiple times, as they are in many restaurants. 

These products which are higher in fat content and lower in carb content have higher amounts of unsaturated fats. Being that the context here is fast food restaurants, the most likely form of unsaturated fat that these would be are polyunsaturated fats from rancid seed oils. 

These products which did show up here had typos or errors in the entry for the fiber content. If the carb content is expanded from net carbs, we can see that these products have 0g of sugar, which is good, and fairly low carbohydrate content, except for item 173 which supposedly has 137.8g of carbs for grilled chicken wings. This must also be an error or typo in the dataset, since grilled chicken does not have anywhere near this amount of carbs.
